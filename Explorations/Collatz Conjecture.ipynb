```python
# Standard Collatz, user input

i = int(input("Type starting value"))

while True:
    if i % 2 == 0:
        print(i, end = " -> ")
        i = int(i/2)
    elif i == 1:
        print(i)
        break
    else:
        print(i, end = " -> ")
        i = i*3+1

```

```python
# Calulating stopping time

import matplotlib.pyplot as plt
maximum = 100000
count_list = []
for j in range(1, maximum):
    count = 0
    i = j
    while True:
        if i % 2 == 0:
            i = int(i/2)
            count += 1
        elif i == 1:
            count += 1
            break
        else:
            i = i*3+1
            count += 1
    count_list.append(count)
plt.plot(range(1, maximum), count_list, "o")
plt.title("Stopping time for Collatz Conjecture")
plt.xlabel("Starting value")
plt.ylabel("Number of iterations");

print(count_list.index(max(count_list)))
```

```python
# Maximum value
maximum = 5000
maximums = []

for i in range(1, maximum):
    temp = []
    while True:
        if i % 2 == 0:
            temp.append(i)
            i = int(i/2)
        elif i == 1:
            temp.append(i)
            break
        else:
            temp.append(i)
            i = i*3+1
    maximums.append(max(temp))
plt.scatter(range(1, maximum), maximums, s=0.5)
plt.ylim([0, 9500])
```
