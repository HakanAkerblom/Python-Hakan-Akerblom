```python
# Standard Collatz, user input

i = int(input("Type starting value"))

while True:
    if i % 2 == 0:
        print(i, end = " -> ")
        i = int(i/2)
    elif i == 1:
        print(i)
        break
    else:
        print(i, end = " -> ")
        i = i*3+1

```

```python
# Calulating stopping time

import matplotlib.pyplot as plt
maximum = 1000
count_list = []
for j in range(1, maximum):
    count = 0
    i = j
    while True:
        if i % 2 == 0:
            i = int(i/2)
            count += 1
        elif i == 1:
            count += 1
            break
        else:
            i = i*3+1
            count += 1
    count_list.append(count)
plt.plot(range(1, maximum), count_list, "o")
plt.title("Stopping time for Collatz Conjecture")
plt.xlabel("Starting value")
plt.ylabel("Number of iterations");
```
