```python
# plotting, inspired by https://github.com/Andreas-Svensson/Python-Andreas-Svensson/blob/main/Laborations/Geometry%20OOP/using_shapes.ipynb

fig, ax = plt.subplots()
r1, r2, r3, c1, c2 = (
    Rectangle(width=2),
    Rectangle(height=2),
    Rectangle(3, -3, 2, 2),
    Circle(x=3),
    Circle(y=-3),
)
shape_list = [r1, r2, r3, c1, c2]

for shape in shape_list:
    ax.add_patch(shape.plot())

ax.autoscale()
ax.set_aspect(1)
```

```python
# plotting translates, inspired by https://github.com/Andreas-Svensson/Python-Andreas-Svensson/blob/main/Laborations/Geometry%20OOP/using_shapes.ipynb

fig, ax = plt.subplots()
r1, r2, r3, c1, c2 = (
    Rectangle(width=2),
    Rectangle(height=2),
    Rectangle(3, -3, 2, 2),
    Circle(x=3),
    Circle(y=-3),
)
shape_list = [r1, r2, c1, c2, r3] # putting r3 last to put it on top of translated circle

for shape in shape_list:
    # create patches of each shape in their pre- and post- translation position
    patch_list = shape.plot_translation(-1, 2)
    for patch in patch_list:
        ax.add_patch(patch) # add specific patch to ax

ax.autoscale()
ax.set_aspect(1)
```
