# while loop

- "if-loop"

```py
while condition:
    statements
    ...
```

````python
i = 0

while i < 10:
    print(f"Hej {i=}")
    i += 1
````

````python
birds = 80000
i = 0
while birds > 8000:
    i += 1
    birds /= 2
print(f"Det tar {i} år")
````

````python
import matplotlib.pyplot as plt

bac = 1.5e6
h = 0

bac_list=[bac]

while bac < 1e7:
    h += 1
    bac *= 1.5
    bac_list.append(bac)
print(f"Det tar {h} timmar")

hours_list = range(0, h+1)
plt.plot(hours_list, bac_list, 'o--')
plt.plot(hours_list, [1e7]*(h+1))
plt.grid()
plt.xlabel("Hours")
plt.ylabel("Number of bacterias")
plt.title("Bacterial growth in milk")
````
