# Functions

```python
# define a function
def cuber(x):
    return x**3

# calls the function cuber using 1 argument
cuber(5)
cubes = [cuber(i) for i in range(-10,10)]
cubes
```

```python
import matplotlib.pyplot as plt

plt.plot(cubes);
```

```python
def my_mean(number1, number2, number3):
    print(f"{number1=}, {number2=}, {number3=}")
    return (number1 + number2 + number3)/3
my_mean(5, 6, 7)
```

## Default values

```python
def hello(number_of_greetings, name = "buddy", age = 0):
    if age < 0:
        print("Ange en positiv ålder")
        return

    for i in range(number_of_greetings):
        print(f"Hej {name}, du är {age} år")
    # note no return -> void function

hello(2, "Håkan", -2)
hello(2, "Håkan", 38)

# name and age are keyword arguments
# number_of_greetings is a positional argument
hello(1, age = 34)
```

```python
def word_counter(text = ""):
    word_list = text.split()
    return len(word_list)

quote = "f f f f f f"
word_counter(quote)
```

## Arbitrary arguments - *args

```python
def my_mean2(*numbers):
    sum_ = 0
    for number in numbers:
        sum_ += number
    return sum_/len(numbers)

my_mean2(1, 2, 3, 4, 5)
```

## Lambda

- anonymous functions

```python
cuber = lambda x: x**3

cuber(3)
```

## Draw lines

```python
def draw_lines(x_list, k=1, m=1, title = "Graf"):
    y_list = [k * x + m for x in x_list]

    plt.plot(x_list, y_list)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)

x = list(range(-10, 10))
draw_lines(x, title = "Lines")
draw_lines(x, k=-2, m=1, title = "Lines")

# TODO find intersection point and plot it
```
