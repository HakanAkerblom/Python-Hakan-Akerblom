# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# array from -10 to 10 with 100 numbers equally spaced between
x = np.linspace(-10, 10, 100)
f = lambda x: np.cos(x)
# <=>
# def f(x):
#   return np.cos(x)
```

## MATLAB style interface

```python


plt.figure(figsize=(8, 4), dpi = 200)
plt.plot(x, f(x), "o--")
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

```python
plt.plot(f(x)) # det går att plotta såhär också, men då får man x-värden från 0 och sedan ökande med 1
```

### OOP Approach

```python
# instantiated fig, ax from figure and axes
fig, ax = plt.figure(figsize = (8, 4)), plt.axes()
ax.plot(x, f(x)) # ax object has a plot method
ax.set(xlabel = "x", ylabel = "y", title = "$\cos{x}$", xlim = (-5,5), ylim = (-1.5, 1.5))
fig.savefig("../assets/cos(x).png", dpi = 200)
```

### Subplots

```python
#MATLAB style
plt.figure(figsize=(8, 4))
plt.subplot(2, 2, 1) # rows, columns, panel number
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("sin(x) and cos(x)")

plt.subplot(2, 2, 2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x**2))


plt.subplot(2, 2, 4)
plt.plot(x, x**2)

plt.suptitle("Trigonometric functions", fontweight = "bold")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 5))

ax[0].plot(x, f(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)")
fig.suptitle("Big title", fontsize = 25)
```
