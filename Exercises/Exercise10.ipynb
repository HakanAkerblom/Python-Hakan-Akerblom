```python
# b)
import matplotlib.pyplot as plt
import numpy as np
means = []

for i in range(1, 8):
    number_of_dices = 10**i
    rolls = np.random.randint(1, 7, number_of_dices)
    mean = np.mean(rolls)
    means.append(mean)
plt.plot(means, "o-")
plt.xticks([i for i in range(7)], [10**i for i in range(1, 8)]);
```
